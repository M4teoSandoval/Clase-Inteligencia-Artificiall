# PCA

```python
# Estandarización de variables numéricas (StandardScaler)

from sklearn.preprocessing import StandardScaler

# Seleccionar columnas numéricas para estandarizar
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
# Excluir 'CUST_ID' si aún está presente
if 'CUST_ID' in numeric_cols:
    numeric_cols.remove('CUST_ID')


# Inicializar el StandardScaler
scaler = StandardScaler()

# Aplicar estandarización a las columnas numéricas seleccionadas
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("\nDataFrame después de la estandarización:")
print(df.head())
```

```python
# Aplicación de PCA. Determine cuántos componentes usando la Varianza Explicada Acumulada del 85%

from sklearn.decomposition import PCA

# Aplicar PCA
pca = PCA()
pca.fit(df[numeric_cols])

# Varianza explicada por cada componente
explained_variance_ratio = pca.explained_variance_ratio_

# Varianza explicada acumulada
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Determinar el número de componentes para alcanzar el 85% de varianza explicada
n_components = np.argmax(cumulative_explained_variance >= 0.85) + 1

print(f"Número de componentes principales para el 85% de varianza explicada: {n_components}")

# Graficar la varianza explicada acumulada
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Varianza Explicada Acumulada por Número de Componentes')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.axhline(y=0.85, color='r', linestyle='-', label='85% de Varianza')
plt.axvline(x=n_components, color='g', linestyle='-', label=f'{n_components} componentes')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Análisis del scree plot (varianza explicada acumulada)

# Crear un DataFrame con los componentes principales seleccionados
pca_final = PCA(n_components=n_components)
df_pca = pca_final.fit_transform(df[numeric_cols])

# Convertir el resultado a un DataFrame de pandas
df_pca = pd.DataFrame(df_pca, columns=[f'PC{i+1}' for i in range(n_components)])

print("\nDataFrame con los componentes principales:")
print(df_pca.head())
```

Se aplicó análisis de componentes principales (PCA) para reducir la dimensionalidad del dataset. El criterio fue retener los componentes que expliquen al menos el 85% de la varianza acumulada.

Según el análisis, se requieren 7 componentes principales para alcanzar ese umbral, lo cual permite simplificar el espacio de características de 17 variables originales a solo 7, manteniendo el 85% de la información estadística del dataset.

Esta reducción mejora la eficiencia computacional y la visualización, además de disminuir el ruido antes de aplicar los modelos de clustering.

# Metodo de el codo

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
sil_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_pca)
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(df_pca, kmeans.labels_))

# Plot elbow and silhouette
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia, marker='o')
plt.title("Método del Codo")
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), sil_scores, marker='o', color='green')
plt.title("Coeficiente de Silueta")
plt.show()
```

El "codo" se observa alrededor de 4 clusters, ya que a partir de ahí la disminución de la inercia empieza a ser menos pronunciada.

El mejor valor de silueta se alcanza con 2 clusters (~0.29), pero este valor suele indicar clusters muy grandes y generales.

Por lo tanto se usaran 4 clusters ya que cumple con la condicion de el cliente (<6), ademas es una decisión equilibrada y técnicamente justificable.

```python
from sklearn.cluster import DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(df_pca)

# Agglomerative Clustering (Ward linkage)
agglo = AgglomerativeClustering(n_clusters=4, linkage='ward')
agglo_labels = agglo.fit_predict(df_pca)

# Gaussian Mixture Models
gmm = GaussianMixture(n_components=4, random_state=42)
gmm_labels = gmm.fit_predict(df_pca)

# Función para evaluar modelos
def evaluar_modelo(nombre, etiquetas):
    if len(set(etiquetas)) <= 1:
        print(f"{nombre}: No se formaron clusters válidos.")
        return
    sil = silhouette_score(df_pca, etiquetas)
    db = davies_bouldin_score(df_pca, etiquetas)
    ch = calinski_harabasz_score(df_pca, etiquetas)
    print(f"{nombre}:\n  Silhouette Score = {sil:.4f}\n  Davies-Bouldin Index = {db:.4f}\n  Calinski-Harabasz Index = {ch:.4f}\n")

# Evaluación
print("🔍 Evaluación de Modelos de Clustering:")
evaluar_modelo("DBSCAN", db_labels)
evaluar_modelo("Agglomerative Clustering", agglo_labels)
evaluar_modelo("Gaussian Mixture Model", gmm_labels)
```
