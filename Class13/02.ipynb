```python
# prompt: crear una grafica de barras con habitaciones comentarios en español

import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV (asumiendo que el código anterior ya lo ha cargado en df)
# Si no está cargado, descomenta la siguiente línea y ajusta la ruta si es necesario
# df = pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv")

# Verifica si la columna 'habitaciones' existe en el DataFrame
if 'habitaciones' in df.columns:
  # Crea la gráfica de barras
  plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura si es necesario
  df['habitaciones'].value_counts().plot(kind='bar', color='skyblue')
  plt.title('Distribución de Habitaciones')
  plt.xlabel('Número de Habitaciones')
  plt.ylabel('Cantidad de Viviendas')
  plt.xticks(rotation=0)  # Evita la rotación de las etiquetas del eje x
  plt.grid(axis='y', alpha=0.5) #Agrega una cuadrícula para facilitar la lectura
  plt.show()
else:
  print("La columna 'habitaciones' no existe en el DataFrame.")
```

```python
# prompt: borrar todos los registros que sean superior a 31 en Area_parqeo y dejarlo en el mismo df

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataframe (assuming it's already loaded as 'df' from the previous code)
# If not loaded, uncomment the following line and adjust the path if necessary
df = pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv")

# Check if the 'Area_parqeo' column exists
if 'Area_parqeo' in df.columns:
  # Filter out rows where 'Area_parqeo' is greater than 31
  df = df[df['Area_parqeo'] <= 31]

  #Now you can use the filtered df for any further operations or visualizations
  plt.figure(figsize=(8, 6))  # Adjust figure size if needed
  df['Area_parqeo'].plot(kind='box', vert=True) # Vertical box plot
  plt.title('Box Plot of Parking Area (After Filtering)')
  plt.ylabel('Area_parqeo')
  plt.show()

else:
  print("The 'Area_parqeo' column does not exist in the DataFrame.")
```

```python
# prompt: crear un arreglo con la variable datos,con datos sinteticos, usando numpy, donde la columna 0 sea la edad y la columna 1 sean las notas de IA que van de 0 a 5.

import numpy as np

# Número de muestras
n_samples = 20

# Crear un arreglo de edades aleatorias entre 18 y 60
edades = np.random.randint(18, 61, size=n_samples)

# Crear un arreglo de notas aleatorias entre 0 y 5 (con un decimal)
notas = np.round(np.random.uniform(0, 5, size=n_samples), 1)

# Combinar los arreglos en un solo arreglo 2D usando numpy
datos = np.column_stack((edades, notas))

datos
```

```python
#Arreglo tipo object
nombres_colombianos = np.array([
    "Juan David", "Maria Camila", "Sebastian", "Sofia", "Daniel", "Valentina",
    "Alejandro", "Laura", "Mateo", "Isabella", "Nicolas", "Camila",
    "Santiago", "Gabriela", "Samuel", "Salome", "David", "Valeria", "Miguel",
    "Luciana"
])

nombres_colombianos
```

```python
# prompt: crear un dataframe llamado alumnos en donde integremos el arreglo nombres_colombianos y edad y notas

# Crear el DataFrame 'alumnos'
alumnos = pd.DataFrame({
    'Nombre': nombres_colombianos,
    'Edad': edades,  # Usar la variable 'edades' creada previamente
    'Notas': notas   # Usar la variable 'notas' creada previamente
})

# Mostrar el DataFrame
alumnos
```

```python
# prompt: calcula el promedio de edad y notas

# Calculate the average age
average_age = alumnos['Edad'].mean()

# Calculate the average grade
average_grade = alumnos['Notas'].mean()

print(f"El promedio de edad es: {average_age}")
print(f"El promedio de notas es: {average_grade}")
```

```python
# prompt: quiero ver las notas de los alumnos que sacaron mas de 3,5

# Filter students with grades above 3.5
high_achievers = alumnos[alumnos['Notas'] > 3.5]

# Display the filtered DataFrame
high_achievers
```
